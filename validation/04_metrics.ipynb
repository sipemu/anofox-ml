# 04 — Metrics

Comparison of sklearn vs rustml evaluation metrics.

```python
import numpy as np
from sklearn.metrics import (
    accuracy_score, confusion_matrix, precision_score, recall_score, f1_score,
    mean_squared_error, mean_absolute_error, r2_score
)
```

## Regression Metrics

```python
y_true = np.array([3.0, -0.5, 2.0, 7.0])
y_pred = np.array([2.5,  0.0, 2.0, 8.0])

print(f'MSE: {mean_squared_error(y_true, y_pred)}')
print(f'MAE: {mean_absolute_error(y_true, y_pred)}')
print(f'R²:  {r2_score(y_true, y_pred)}')
```

## Binary Classification Metrics

```python
y_true_cls = np.array([0, 0, 1, 1, 0, 1, 1, 0])
y_pred_cls = np.array([0, 1, 1, 1, 0, 0, 1, 0])

print(f'Accuracy:  {accuracy_score(y_true_cls, y_pred_cls)}')
print(f'Confusion: \n{confusion_matrix(y_true_cls, y_pred_cls)}')
print(f'Precision: {precision_score(y_true_cls, y_pred_cls, average=None)}')
print(f'Recall:    {recall_score(y_true_cls, y_pred_cls, average=None)}')
print(f'F1:        {f1_score(y_true_cls, y_pred_cls, average=None)}')
```

## Multiclass Classification

```python
y_true_m = np.array([0, 1, 2, 0, 1, 2, 0, 1, 2])
y_pred_m = np.array([0, 2, 1, 0, 0, 2, 0, 1, 2])

print(f'Accuracy:  {accuracy_score(y_true_m, y_pred_m)}')
print(f'Confusion: \n{confusion_matrix(y_true_m, y_pred_m)}')
print(f'Precision: {precision_score(y_true_m, y_pred_m, average=None)}')
print(f'Recall:    {recall_score(y_true_m, y_pred_m, average=None)}')
print(f'F1:        {f1_score(y_true_m, y_pred_m, average=None)}')
```
