# 03 — Decision Trees

Side-by-side comparison of sklearn vs rustml decision tree classifier and regressor.

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
```

## Decision Tree Classifier

```python
X_train = np.array([
    [1.0, 2.0], [2.0, 1.0], [2.0, 3.0], [3.0, 2.0],
    [7.0, 8.0], [8.0, 7.0], [8.0, 9.0], [9.0, 8.0]
])
y_train = np.array([0, 0, 0, 0, 1, 1, 1, 1], dtype=float)

for depth in [None, 2, 1]:
    clf = DecisionTreeClassifier(max_depth=depth, random_state=42)
    clf.fit(X_train, y_train)
    X_test = np.array([[1.5, 1.5], [8.5, 8.5], [5.0, 5.0]])
    print(f'depth={depth}: preds={clf.predict(X_test)}, importances={clf.feature_importances_}')
```

## Tree Visualization

```python
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X_train, y_train)

plt.figure(figsize=(12, 6))
plot_tree(clf, filled=True, feature_names=['x₁', 'x₂'], class_names=['0', '1'])
plt.title('Decision Tree (max_depth=3)')
plt.show()
```

## Decision Tree Regressor

```python
X_reg = np.array([[1], [2], [3], [4], [5], [6], [7], [8]], dtype=float)
y_reg = np.array([2.1, 3.9, 6.2, 7.8, 10.1, 12.0, 13.9, 16.1])
X_test_reg = np.array([[1.5], [4.5], [7.5]])

for depth in [None, 2, 1]:
    reg = DecisionTreeRegressor(max_depth=depth, random_state=42)
    reg.fit(X_reg, y_reg)
    print(f'depth={depth}: preds={reg.predict(X_test_reg)}, importances={reg.feature_importances_}')
```

## Validation

Golden data in `tests/golden_data/decision_tree.json` covers 7 cases (classifier with gini/entropy,
regressor, varying max_depth). Predictions match sklearn exactly; feature importances match
within 0.15 tolerance (sklearn uses `random_state` for tie-breaking which can shift importances slightly).
