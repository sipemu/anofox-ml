# 02 — K-Nearest Neighbors

Side-by-side comparison of sklearn vs rustml KNN classifier and regressor.

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
import matplotlib.pyplot as plt
```

## KNN Classifier — Two Clusters

```python
X_train = np.array([
    [0.0, 0.0], [0.5, 0.5], [1.0, 0.0], [0.0, 1.0],
    [5.0, 5.0], [5.5, 5.5], [6.0, 5.0], [5.0, 6.0]
])
y_train = np.array([0, 0, 0, 0, 1, 1, 1, 1])
X_test = np.array([[0.2, 0.3], [5.2, 5.3], [2.5, 2.5]])

for k in [3, 5]:
    for w in ['uniform', 'distance']:
        clf = KNeighborsClassifier(n_neighbors=k, weights=w)
        clf.fit(X_train, y_train)
        print(f'k={k}, weights={w}: {clf.predict(X_test)}')
```

## KNN Regressor — Linear Data

```python
X_reg = np.array([[1], [2], [3], [4], [5], [6], [7], [8]], dtype=float)
y_reg = np.array([2, 4, 6, 8, 10, 12, 14, 16], dtype=float)
X_test_reg = np.array([[2.5], [5.5], [1.0]])

for k in [3, 5]:
    for w in ['uniform', 'distance']:
        reg = KNeighborsRegressor(n_neighbors=k, weights=w)
        reg.fit(X_reg, y_reg)
        print(f'k={k}, weights={w}: {reg.predict(X_test_reg)}')
```

## Decision Boundary Visualization

```python
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)

xx, yy = np.meshgrid(np.linspace(-1, 7, 200), np.linspace(-1, 7, 200))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='RdBu')
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='RdBu', edgecolors='k', s=80)
plt.title('KNN (k=3) Decision Boundary')
plt.xlabel('x₁')
plt.ylabel('x₂')
plt.show()
```
